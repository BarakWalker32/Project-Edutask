```python
# Cell 1: Install ipywidgets
!pip install ipywidgets

# Cell 2: Import required libraries
import ipywidgets as widgets
from IPython.display import display
import os
import sqlite3
import hashlib

# Cell 3: Define UserAuthentication Class
class UserAuthentication:
    def __init__(self):
        pass  # No need for initialization for now

    def register_user(self):
        pass  # No need for registration for now

    def login_user(self):
        pass  # No need for login for now

    def logout_user(self):
        pass  # No need for logout for now

# Cell 4: Define Class ToDo Planner
class ToDoPlanner:
    def __init__(self):
        self.task_input = widgets.Text(placeholder='Enter task...')
        self.add_button = widgets.Button(description='Add Task')
        self.add_button.on_click(self.add_task)
        self.task_list = widgets.VBox()  # Initialize task_list
        self.container = widgets.VBox([self.task_input, self.add_button, self.task_list])
        self.load_tasks()  # Load tasks after task_list is initialized

    def load_tasks(self):
        self.tasks = {}
        tasks_file = os.path.join(os.getcwd(), 'tasks.txt')
        try:
            with open(tasks_file, 'r') as f:
                for line in f:
                    task = line.strip()
                    if task:
                        self.tasks[task] = widgets.HBox([widgets.Checkbox(description=task), widgets.Button(description='Delete', layout=widgets.Layout(margin='5px 0 0 10px'))])
                        self.tasks[task].children[1].on_click(self.delete_task)
                        self.task_list.children += (self.tasks[task],)
        except FileNotFoundError:
            pass

    def save_tasks(self):
        tasks_file = os.path.join(os.getcwd(), 'tasks.txt')
        with open(tasks_file, 'w') as f:
            for task in self.tasks:
                f.write(task + '\n')

    def add_task(self, b):
        task = self.task_input.value.strip()
        if task:
            if task not in self.tasks:
                self.tasks[task] = widgets.HBox([widgets.Checkbox(description=task), widgets.Button(description='Delete', layout=widgets.Layout(margin='5px 0 0 10px'))])
                self.tasks[task].children[1].on_click(self.delete_task)
                self.task_list.children += (self.tasks[task],)
                self.task_input.value = ''
                self.save_tasks()
            else:
                print("Task already exists!")
        else:
            print("Task cannot be empty!")

    def delete_task(self, b):
        for task, widget in self.tasks.items():
            if b == widget.children[1]:
                del self.tasks[task]
                self.task_list.children = tuple([item for item in self.task_list.children if item != widget])
                self.save_tasks()
                break

    def display(self):
        display(self.container)

# Cell 5: Display menu and ToDoPlanner
def display_menu():
    print("Welcome, JSU student!")
    print("Instructions:")
    print("- Enter a task in the text box below and click 'Add Task' to add it to your to-do list.")
    print("- Check the checkbox next to a task to mark it as completed.")
    print("- Click the 'Delete' button next to a task to remove it from your list.")

# Creating the UserAuthentication instance
authenticator = UserAuthentication()

# If the user successfully logs in (skipping for now), display the menu and ToDoPlanner
if True:  # Skipping the login process for now
    display_menu()
    planner = ToDoPlanner()
    planner.display()
```
